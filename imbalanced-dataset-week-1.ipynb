```python
import numpy as np
import pandas as pd
df = pd.read_csv("train.csv")
df
```

```python
y = np.array(df["Survived"])
df = df[["Pclass", "Age", "Sex", "Fare"]]
df.isna().sum()
x = df["Age"]
x.dropna(inplace=True)
median = np.median(np.array(x))
df.replace(np.nan, median, inplace=True)
df.isna().sum()
```

```python
df.replace(["male", "female"], [0, 1], inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split
df = np.array(df)
X_train, x_test, y_train, y_test = train_test_split(df, y, test_size=0.25, random_state=42)
```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.naive_bayes import GaussianNB

nb_classifier = GaussianNB()
kf = KFold(n_splits=10, shuffle=True, random_state=42)


def kfoldmetrics(model):
    accuracy_scores = cross_val_score(
        model, X_train, y_train, cv=kf, scoring="accuracy"
    )
    for i, score in enumerate(accuracy_scores, 1):
        print(f"Fold {i}: Accuracy = {score:.4f}")

    # Print the average accuracy across all folds
    print(f"\nAverage Accuracy: {accuracy_scores.mean():.4f}")

kfoldmetrics(nb_classifier)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5, p=2)
kfoldmetrics(knn)
```

```python
from sklearn.linear_model import LogisticRegression

logisticReg = LogisticRegression()
kfoldmetrics(logisticReg)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(
    n_estimators=100, max_features=4, random_state=42
)

kfoldmetrics(rf_classifier)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

dt_classifier = DecisionTreeClassifier(random_state=42)

# Create the Bagging classifier with 100 trees
bagging_classifier = BaggingClassifier(
    estimator=dt_classifier, n_estimators=100, random_state=42
)

kfoldmetrics(bagging_classifier)
```

```python
from sklearn.ensemble import ExtraTreesClassifier

et_classifier = ExtraTreesClassifier(n_estimators=100, max_features=4, random_state=42)
kfoldmetrics(et_classifier)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

# Create the base Decision Tree classifier for AdaBoost
base_classifier = DecisionTreeClassifier(random_state=42)

# Create the AdaBoost classifier with 50 trees
adaboost_classifier = AdaBoostClassifier(
    base_classifier, n_estimators=50, random_state=42
)

kfoldmetrics(adaboost_classifier)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_classifier = GradientBoostingClassifier(n_estimators=50, random_state=42)

kfoldmetrics(gb_classifier)
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

logreg_classifier = LogisticRegression()
dt_classifier = DecisionTreeClassifier(random_state=42)
svm_classifier = SVC(probability=True)  # Set probability=True for soft voting
nb_classifier = GaussianNB()
knn_classifier = KNeighborsClassifier(n_neighbors=5, p=2)


ensemble_classifier = VotingClassifier(
    estimators=[
        ("logreg", logreg_classifier),
        ("dt", dt_classifier),
        ("svm", svm_classifier),
        ("nb", nb_classifier),
        ("knn", knn_classifier),
    ],
    voting="soft",
    weights=[1, 2, 1, 2, 3],
)

kfoldmetrics(ensemble_classifier)
```

```python
df = pd.read_csv("bank.csv")
df
```

```python
y = df["y"]
df.drop("y", axis=1, inplace=True)
df.isna().sum()
```

```python
df.value_counts()
```

```python
df.replace(["yes", "no"], [1, 0], inplace=True)
df_encoded = pd.get_dummies(df, columns=["job", "marital", "education", "contact", "poutcome"])
df_encoded
```

```python
df_encoded.replace(["False", "True"], [0, 1], inplace=True)
month_dict = {
    "jan": 1,
    "feb": 2,
    "mar": 3,
    "apr": 4,
    "may": 5,
    "jun": 6,
    "jul": 7,
    "aug": 8,
    "sep": 9,
    "oct": 10,
    "nov": 11,
    "dec": 12,
}


df_encoded.replace(month_dict, inplace=True)
y.replace(["yes", "no"], [1, 0], inplace=True)
y
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(df_encoded, y, test_size=0.25, random_state=42)
print(xtrain.shape, xtest.shape, ytrain.shape, ytest.shape)
```

```python
def kfoldmetrics1(model):
    accuracy_scores = cross_val_score(
        model, xtrain, ytrain, cv=kf, scoring="accuracy"
    )
    for i, score in enumerate(accuracy_scores, 1):
        print(f"Fold {i}: Accuracy = {score:.4f}")

    # Print the average accuracy across all folds
    print(f"\nAverage Accuracy: {accuracy_scores.mean():.4f}")
```
